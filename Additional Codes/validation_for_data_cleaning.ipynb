```
print(text1)
print(len(text1))
```

```
# Removing Lablling phrases
for phrase in label_phrases:
    if phrase in text1.lower():
      print(phrase)
      text1=text1.replace(phrase,'')
print()
print(text1)
print(len(text1))
```

```
# Removing enitities using spacy that are not useful
nlp = en_core_web_sm.load()
text2= nlp(text1)
displacy.render(text2, style = "ent",jupyter = True)
entities=[(i, i.label_, i.label) for i in text2.ents]
# print(entities)
for x,y,z in entities:
  if y=='DATE' or y=='PERSON' or y=='GPE' or y=='ORG':
    # print(str(x))
    text1=text1.replace(str(x),'')
```

```
print(text1)
print(len(text1))
```

```
# Removing punctuation
remove = string.punctuation
remove = remove.replace("/", "")
text1=text1.translate(str.maketrans('', '', remove))
```
